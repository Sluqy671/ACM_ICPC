```
# initializaion

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn import metrics
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

```

##**Part.a**  

Read in "Advertising" data.

```
url = 'http://faculty.marshall.usc.edu/gareth-james/ISL/Advertising.csv'
df = pd.read_csv(url, index_col = 0)
df.head()
```

Though for radio and newspaper it is hard to learn the clear relationships between them and sales.

&nbsp;&nbsp;&nbsp;&nbsp;Then, split the data into training data (70%) and test data (30%).




```
train, test = train_test_split(df, test_size = 0.3)
train.info()
test.info()
```

##**Part.b**
Use the training data to fit a simple (univariate) linear regression to predict product sales, for each of three features: TV ad budget, radio ad budget, and newspaper ad budget.  
For each of the three regression models, create a plot of the data and the regression line, with
product sales (y) on the vertical axis and the feature on which the model was trained (x) on
the horizontal axis.  
Also, for each regression model, print
the intercept and coefficients, and compute the MSE and R2 on the training data, and MSE
and R2 on the test data.

**Linear Regression**

```
reg_tv = LinearRegression().fit(train[['TV']], train['sales'])
reg_radio = LinearRegression().fit(train[['radio']], train['sales'])
reg_news = LinearRegression().fit(train[['newspaper']], train['sales'])
```

**Plot data and regression line**

```
fig = plt.figure(figsize=(12,3))

plt.subplot(1,3,1)
sns.scatterplot(data=train, x="TV", y="sales");
sns.lineplot(data=train, x="TV", y=reg_tv.predict(train[['TV']]), color='red');

plt.subplot(1,3,2)
sns.scatterplot(data=train, x="radio", y="sales");
sns.lineplot(data=train, x="radio", y=reg_radio.predict(train[['radio']]), color='red');

plt.subplot(1,3,3)
sns.scatterplot(data=train, x="newspaper", y="sales");
sns.lineplot(data=train, x="newspaper", y=reg_news.predict(train[['newspaper']]), color='red');
```

**Intercept and coefficients**

```
print("TV: ", reg_tv.coef_[0], reg_tv.intercept_)
print("Radio: ", reg_radio.coef_[0], reg_radio.intercept_)
print("Newspaper: ", reg_news.coef_[0], reg_news.intercept_)
```

&nbsp;&nbsp;&nbsp;&nbsp;From the calculated coefficients, radio ad budget seems to have the most powerful effect on sales, followed by TV and then by newspaper.

**MSE and R2 on the training data**

```
y_tv_hat = reg_tv.predict(train[['TV']])
y_radio_hat = reg_radio.predict(train[['radio']])
y_news_hat = reg_news.predict(train[['newspaper']])
mse_tv = metrics.mean_squared_error(train[['sales']], y_tv_hat)
mse_radio = metrics.mean_squared_error(train[['sales']], y_radio_hat)
mse_news = metrics.mean_squared_error(train[['sales']], y_news_hat)
print("MSE_TV: ", mse_tv)
print("MSE_Radio: ", mse_radio)
print("MSE_Newspaper: ", mse_news)
```

```
r2_tv = 1-np.mean( (y_tv_hat - train['sales'])**2 / np.std(train['sales'])**2 )
r2_radio = 1-np.mean( (y_radio_hat - train['sales'])**2 / np.std(train['sales'])**2 )
r2_news = 1-np.mean( (y_news_hat - train['sales'])**2 / np.std(train['sales'])**2 )
print("R2_TV: ", r2_tv)
print("R2_Radio: ", r2_radio)
print("R2_Newspaper: ", r2_news)
```

**MSE and R2 on the test data**

```
test_y_tv_hat = reg_tv.predict(test[['TV']])
test_y_radio_hat = reg_radio.predict(test[['radio']])
test_y_news_hat = reg_news.predict(test[['newspaper']])
test_mse_tv = metrics.mean_squared_error(test[['sales']], test_y_tv_hat)
test_mse_radio = metrics.mean_squared_error(test[['sales']], test_y_radio_hat)
test_mse_news = metrics.mean_squared_error(test[['sales']], test_y_news_hat)
print("MSE_TV: ", test_mse_tv)
print("MSE_Radio: ", test_mse_radio)
print("MSE_Newspaper: ", test_mse_news)
```

```
test_r2_tv = 1-np.mean( (test_y_tv_hat - test['sales'])**2 / np.std(test['sales'])**2 )
test_r2_radio = 1-np.mean( (test_y_radio_hat - test['sales'])**2 / np.std(test['sales'])**2 )
test_r2_news = 1-np.mean( (test_y_news_hat - test['sales'])**2 / np.std(test['sales'])**2 )
print("R2_TV: ", test_r2_tv)
print("R2_Radio: ", test_r2_radio)
print("R2_Newspaper: ", test_r2_news)
```

&nbsp;&nbsp;&nbsp;&nbsp;According to the calculation results, in both training and test data, the value of MSE of TV model is the smallest, and the value of  R2 of TV model is the largest, so TV ad budget model is most effective at predicting product sales.

```
#prediction test_y_tv_hat
#real test['sales']

fig = plt.figure(figsize=(12,3))

plt.subplot(1, 3, 1)
plt.xlabel('actual sales(y)')
plt.ylabel('TV-predicted sales(y^)')
plt.xlim(5, 30)
plt.ylim(5, 30)
plt.plot(test['sales'], test_y_tv_hat, 'o')

plt.subplot(1, 3, 2)
plt.xlabel('actual sales(y)')
plt.ylabel('Radio-predicted sales(y^)')
plt.xlim(5, 30)
plt.ylim(5, 30)
plt.plot(test['sales'], test_y_radio_hat, 'o')

plt.subplot(1, 3, 3)
plt.xlabel('actual sales(y)')
plt.ylabel('News-predicted sales(y^)')
plt.xlim(5, 30)
plt.ylim(5, 30)
plt.plot(test['sales'], test_y_news_hat, 'o')
```

### * For each of the three regression models, compute the residuals (y - yˆ). Note that some of these will be negative, and some will be positive.

```
tv_residual = test['sales'] - test_y_tv_hat
radio_residual = test['sales'] - test_y_radio_hat
news_residual = test['sales'] - test_y_news_hat

tv_residual = np.array(tv_residual)
radio_residual = np.array(radio_residual)
news_residual = np.array(news_residual)

mr_tv = np.mean(tv_residual)
mr_radio = np.mean(radio_residual)
mr_news = np.mean(news_residual)

print("mean residual of TV: ", mr_tv)
print("mean residual of radio: ", mr_radio)
print("mean residual of news: ", mr_news)
```

&nbsp;&nbsp;&nbsp;&nbsp;For a well-fitted regression model, the mean residual should be close to 0, which means the value of residuals are distributed by the zero line almost equally and no bias on either side.  
&nbsp;&nbsp;&nbsp;&nbsp;But the mean residual near 0 doesn't mean it is a good regression model, the MSE and R2 should also be taken into consideration to determine the goodness of the model.

##**Part.d**

&nbsp;&nbsp;&nbsp;&nbsp;Next, fit a multiple linear regression to predict product sales, using all three features to train a single model: TV ad budget, radio ad budget, and newspaper ad budget. Print the intercept and coefficients, and compute the MSE and R2 on the training data, and MSE and R2 on the test data. Comment on the results.

**Intercept and coefficients**

```
reg_multi_ad = LinearRegression().fit(train[['TV', 'radio', 'newspaper']], train['sales'])
print("Coefficients (TV, radio, newspaper):", reg_multi_ad.coef_)
print("Intercept: ", reg_multi_ad.intercept_)
```

&nbsp;&nbsp;&nbsp;&nbsp;The coefficeints of TV, radio and newspaper ads are different from what they had been in the single regression case. While radio ads still affect a lot on sales, followed by TV ads, newspaper ads seems to have nearly no effects on sales.  
&nbsp;&nbsp;&nbsp;&nbsp;In the simple regression case, the coefficient for newspaper ads represents the effect of an increase in newspaper advertising. But in the multiple regression case, the coefficient for newspaper ads represents the effect of the increase in newspaper advertising while keep TV and radio advertising stay.

**MSE and R2 on the training data**

```
y_train_multi_ad = reg_multi_ad.predict(train[['TV', 'radio', 'newspaper']])

mse_train_multi = metrics.mean_squared_error(train[['sales']], y_train_multi_ad)

r2_train_multi_ad = 1-np.mean( (y_train_multi_ad - train['sales'])**2 / np.std(train['sales'])**2 )

print("MSE_multi: ", mse_train_multi)
print("R2_multi: ", r2_train_multi_ad)
```

&nbsp;&nbsp;&nbsp;&nbsp;When compared with the single regression models before, the multiple regression model now looks much more reliable, for its MSE in the training data is quite smaller and R2 value really close to 1.

**MSE and R2 on the test data**

```
y_test_multi_ad = reg_multi_ad.predict(test[['TV', 'radio', 'newspaper']])

mse_test_multi = metrics.mean_squared_error(test[['sales']], y_test_multi_ad)

r2_test_multi_ad = 1-np.mean( (y_test_multi_ad - test['sales'])**2 / np.std(test['sales'])**2 )

print("MSE_test_multi: ", mse_test_multi)
print("R2_test_multi: ", r2_test_multi_ad)
```

&nbsp;&nbsp;&nbsp;&nbsp;Prediction results from the multiple regression model also appear to be better than the single regression models. Although the MSE_test_multi is a little larger than MSE in training data, and R2 smaller than it in training data, both of them are still good enough for a reliable regression model.  
&nbsp;&nbsp;&nbsp;&nbsp;The calculation results show that there are inner relationships between the features, and a multiple regression model should be better than a single regression model here.

###* Plot predicted sales (yˆ) on the vertical axis, and actual sales (y) on the horizontal axis. Make sure both axes use the same scale.

```
plt.xlabel('actual sales(y)')
plt.ylabel('multi-predicted sales(y^)')
plt.xlim(5, 30)
plt.ylim(5, 30)
plt.plot(test['sales'], y_test_multi_ad, 'o')
```

&nbsp;&nbsp;&nbsp;&nbsp;This plot explains the data very well. As we have discussed before, for an ideal model the predicted value should be close or equal to the true value, which means it is close to the function f(x) = y. The slope is 1 and it starts on (0, 0).

&nbsp;&nbsp;&nbsp;&nbsp;For a well-fitted regression model, the mean of the residuals should be a number near or equal 0. It means the residuals distribute evenly and no bias to be positive or negative.

##**Part.e**

Our multiple linear regression includes additive effects of all three types of advertising media. However, it does not include interaction effects, in which combining different types of
advertising media together results in a bigger boost in sales than just the additive effect of the
individual media. The pattern in the residuals plots from parts (a) through (d) suggest that a
model including an interaction effect may explain sales data better than a model including
additive effects. Add four columns to your data frame: newspaper × radio, TV × radio,
newspaper × TV, and newspaper × radio × TV. Then, train a linear regression model
on all seven features: the three types of ad budgets, and the four interaction effects. Repeat
the analysis of part (d) for the model including interaction effects.

###**Add columns to data frame**

```
n_r = df['newspaper'] * df['radio']
t_r = df['TV'] * df['radio']
n_t = df['newspaper'] * df['TV']
n_r_t = df['newspaper'] * df['radio'] * df['TV']

df['n_r'] = n_r
df['t_r'] = t_r
df['n_t'] = n_t
df['n_r_t'] = n_r_t

df.head()
```

```
train, test = train_test_split(df, test_size = 0.3)
train.info()
test.info()
```

###**Train the multiple regression model**

```
reg_multi_ad = LinearRegression().fit(train[['TV', 'radio', 'newspaper', 'n_r', 't_r', 'n_t', 'n_r_t']], train['sales'])
print("Coefficients (TV, radio, newspaper, n_r, t_r, n_t, n_r_t'):", reg_multi_ad.coef_)
print("Intercept: ", reg_multi_ad.intercept_)
```

###**MSE and R2 on the training data**

```
y_train_multi = reg_multi_ad.predict(train[['TV', 'radio', 'newspaper', 'n_r', 't_r', 'n_t', 'n_r_t']])

mse_train_multi = metrics.mean_squared_error(train[['sales']], y_train_multi)

r2_train_multi = 1-np.mean( (y_train_multi - train['sales'])**2 / np.std(train['sales'])**2 )

print("MSE_multi: ", mse_train_multi)
print("R2_multi: ", r2_train_multi)
```

###**MSE and R2 on the test data**

```
y_test_multi = reg_multi_ad.predict(test[['TV', 'radio', 'newspaper', 'n_r', 't_r', 'n_t', 'n_r_t']])

mse_test_multi = metrics.mean_squared_error(test[['sales']], y_test_multi)

r2_test_multi = 1-np.mean( (y_test_multi - test['sales'])**2 / np.std(test['sales'])**2 )

print("MSE_multi: ", mse_test_multi)
print("R2_multi: ", r2_test_multi)
```

&nbsp;&nbsp;&nbsp;&nbsp;Compared with the multiple regression model before, this time the model behaves better with lower MSE and larger R2.

###**Predicted sales and actual sales**

```
plt.xlabel('actual sales(y)')
plt.ylabel('multi-predicted sales(y^)')
plt.xlim(5, 30)
plt.ylim(5, 30)
plt.plot(test['sales'], y_test_multi, 'o')
```

&nbsp;&nbsp;&nbsp;&nbsp;Pattern here looks like a straight line more. The points converge on the exact line.

###**Calculate residuals**

```
multi_residual = test['sales'] - y_test_multi
multi_residual = np.array(multi_residual)
mr_multi = np.mean(multi_residual)
print("Mean of the residuals: ", mr_multi)
```

&nbsp;&nbsp;&nbsp;&nbsp;Much closer to 0 than before.

###**Plot residuals**

```
plt.xlabel('actual sales(y)')
plt.ylabel('residuals')
plt.plot(test['sales'], multi_residual, 'o')
```

&nbsp;&nbsp;&nbsp;&nbsp;This time the plot is quite different from the plot. However, though the pattern seems different, the main attribute is the same: most points distribute near the line y = 0.  
&nbsp;&nbsp;&nbsp;&nbsp;Another different thing is the left part of the plot. In the plot of multiple regression model with only three features, there is only one single point far from the line y = 0. But this time there are several points. We may come to a hypothesis that when the number of sales is small, the predicted value may be larger than the true value.

###**Plot relationships between features and residuals**

```
fig = plt.figure(figsize = (12, 9))

plt.subplots_adjust(wspace=0.3, hspace=0.5)

plt.subplot(3, 3, 1)
plt.xlabel('TV ad budgets')
plt.ylabel('residuals')
plt.plot(test['TV'], multi_residual, 'o')

plt.subplot(3, 3, 2)
plt.xlabel('radio ad budgets')
plt.ylabel('residuals')
plt.plot(test['radio'], multi_residual, 'o')

plt.subplot(3, 3, 3)
plt.xlabel('newspaper ad budgets')
plt.ylabel('residuals')
plt.plot(test['newspaper'], multi_residual, 'o')

plt.subplot(3, 3, 4)
plt.xlabel('newspaper × radio')
plt.ylabel('residuals')
plt.plot(test['n_r'], multi_residual, 'o')

plt.subplot(3, 3, 5)
plt.xlabel('TV × radio')
plt.ylabel('residuals')
plt.plot(test['t_r'], multi_residual, 'o')

plt.subplot(3, 3, 6)
plt.xlabel('newspaper × TV')
plt.ylabel('residuals')
plt.plot(test['n_t'], multi_residual, 'o')

plt.subplot(3, 3, 7)
plt.xlabel('newspaper × radio × TV')
plt.ylabel('residuals')
plt.plot(test['n_r_t'], multi_residual, 'o')
```

&nbsp;&nbsp;&nbsp;&nbsp;For the first row, the subplots of radio and newspaper appear no much difference from the subplots before. But for TV ad budgets, when the amount is smaller than about 30, the prediction appears to have a larger number than the real sales.  
&nbsp;&nbsp;&nbsp;&nbsp;For the second row, the subplots with TV has the same pattern as the first subplot, which is the affection from TV ad budgets, while for the 4th subplot, the combination of newspaper and radio may have the same affection as TV has. The last subplot is the combination of all three features, and it has then same pattern.
